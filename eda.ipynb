```python
import pandas
data = pandas.read_csv("datasets/supermarket_data.csv")
# show output 
data
```

```python
# Describe data 
data.describe()
```

```python
# Plot a count plot 
import seaborn
seaborn.countplot(data=data,x="gender",palette="Oranges" )
```

```python
# Gender count plot with hue of city 
seaborn.countplot(data=data, x="gender", palette="Reds",hue="city")
```

```python
# Plot a count plot of Column  product without hue
seaborn.countplot(data=data , x="product", palette="Blues" )
import matplotlib.pyplot as ell
ell.xticks(rotation = 35)
```

```python
# Plot a pie chat 
# on column payment
data["payment"].value_counts().plot.pie(autopct ="%1.11f%%")
```
